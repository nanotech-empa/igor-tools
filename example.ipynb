```python
import numpy as np
import matplotlib.pyplot as plt

import igor_tools
```

```python
# Read an .itx file containing multiple waves
waves = igor_tools.read_itx("example_files/waves_1d_and_2d.itx")
waves
```

```python
plt.plot(waves[0].x_arr, waves[0].data)
plt.xlabel(f"{waves[0].x_symbol} {waves[0].x_unit}")
plt.show()
```

```python
plt.imshow(waves[1].data.T, origin='lower', extent=waves[1].extent)
plt.xlabel(f"{waves[1].x_symbol} {waves[1].x_unit}")
plt.ylabel(f"{waves[1].y_symbol} {waves[1].y_unit}")
plt.show()
```

```python
# Writing an igor file:

fname = "example_files/example_out.itx"

igor_tools.write_2d_itx(
    fname,
    np.ones((6, 7)),
    (0.0, 0.1, "m"),
    (0.0, 0.2, "m"),
    "example_output"
)
```

```python
with open(fname, 'r') as f:
    print(f.read())
```
